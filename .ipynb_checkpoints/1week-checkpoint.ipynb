```python
import pandas as pd
import numpy as np
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

```python
df = pd.read_csv("archive/Combined_dataset.csv")
print(f"Dataset loaded: {df.shape[0]} rows × {df.shape[1]} columns")
```

```python
#2.Data Exploration
```

```python
# Displaying first 5 rows
print("FIRST 5 ROWS")
df.head()
```

```python
# Displaying last 5 rows
print("LAST 5 ROWS")
df.tail()
```

```python
#Shape ,rows,colums
print("Shape:", df.shape)
print("Columns:", df.columns)
```

```python
#Displaying data types
print("\nData Types:")
print(df.dtypes)
```

```python

#3:Handling Missing Values

```

```python
#Identifying missing values
print("Missing values per column")
print(df.isnull().sum())
print()
print(f"Total missing values in dataset: {df.isnull().sum().sum()}")
```

```python
#Filling missing values
#Numeric columns with median value
for col in df.select_dtypes(include='number').columns:
    df[col] = df[col].fillna(df[col].median())
    print(f"Filled numeric column '{col}' with median")

#Text/object columns with "Unknown"
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna('Unknown')
    print(f"Filled text column '{col}' with 'Unknown'")
```

```python
#Droping the rows containg missing values
df.dropna(inplace=True)
```

```python
#Verifying if missing values are remaining
print(f"Missing values remaining: {df.isnull().sum().sum()}")
print("No missing values found")
```

```python
#4 BASIC OPERATIONS
```

```python
#Filtering rows
#Check column names
print(df.columns.tolist())
```

```python
# Filtering products with rating more than 4
filtered = df[df['rating'] > 4]
print("High rated products : {len(filtered)}")
filtered.head()
```

```python
#Select columns
#Selecing few important columns
selected = df[['title', 'category', 'rating', 'final_price']]
print("Shape : {selected.shape}")
selected.head()
```

```python
#5. Removing duplicates
```

```python
#Removing duplicates
df.drop_duplicates(inplace=True)
```

```python
# Counting duplicates
print("Duplicate Rows:", df.duplicated().sum())
```

```python
#CREATING DERIVED COLUMNS
```

```python
# Adding a quantity column
df['quantity'] = 1
# Creating total_amount column
df['total_amount'] = df['final_price'] * df['quantity']
# Verifying
print("total_amount COLUMN CREATED")
print(df[['title', 'initial_price', 'discount', 'final_price', 'quantity', 'total_amount']].head(10))
print()
print("Max total amount  : {df['total_amount'].max()}")
print("Min total amount  : {df['total_amount'].min()}")
print("Mean total amount : {df['total_amount'].mean():.2f}")
print("Derived column created!")
```

```python
#Saving the cleaned dataset
df.to_csv("shopping_cleaned.csv", index=False)
```

```python
#Verification
print(f"File name   : shopping_cleaned.csv")
print(f"Final shape : {df.shape[0]} rows × {df.shape[1]} columns")
```
